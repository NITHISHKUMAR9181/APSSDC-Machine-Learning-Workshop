<h1>Dimensionality Reduction:</h1>There are More NO of Columns in the Dataset.It may cause hard to find.

In Dimensionality Reduction we do:

1.Feature Elimination:Here Removing Unnecessary features from the dataset and only using important features.

-->Here not necessary columns are removed totally.

2.Feature Extraction :Transforming 'M' Number of feature into N number of columns--> N<M.

-->Here n columns are combined to form 1 column with smilar features.

EX:Dataset contains 100 columns---->4 Columns.

```python
import pandas as pd
```

```python
product = [1,3,5,6,7,8]

price = [100,50,30,50,60,70]

df = pd.DataFrame({'Product':product,'price':price})
```

```python
df #Here we remove the product Feature  and price feature is necessary ..
```

```python
df['Newcolumn'] = df['Product'] * df['price'] #Feature Extraction .
```

```python
df
```

Algorithm for Feature Extraction.

<h2>Principal Component Analysis(PCA):</h2> It is Technique for Feature Extraction.

WHen we use this:

1.If we want to reduce Number of features but you are not able to identify which feature not necessary to remove.

Now we can apply PCA for Breast Cancer Dataset:

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
%matplotlib notebook
```

```python
#The Above code  % ........... MAgical Function.
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data['data'],columns = data['feature_names'])
```

```python
df['target'] = data['target']
```

```python
df
```

```python
df.shape #30 features ,1 -->Target column
```

```python
X = df.drop('target',axis=1)
X.shape
```

```python
y = df['target']

y.shape
```

```python
# Removing or Extracting the Features. # Standardisation data.
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
#1.Creating Objcet  of scaler obj.

scaler = StandardScaler()
```

```python
#Transorming the Data usinf Fit_tansform.

X_transformed =scaler.fit_transform (X)
```

```python
X_transformed.shape
```

```python
X.describe()
```

```python
pd.DataFrame(X_transformed).describe() #observe the changes after scaling.
```

```python
#Compare the Two columns of the describe().

#Here High Dimensional data is changed to Low Dimensional data.
```

# Importing PCA.

```python
from sklearn.decomposition import PCA
```

```python
#Creating Object #Using to make n_components =2 .

pcaObject = PCA(n_components=2) #Here making 30 columns as 2 principal components.
```

```python
X_pca = pcaObject.fit_transform(X_transformed) 

#X_transformed data is Scaled data.
```

```python
#Making Dataframe .

df_pca = pd.DataFrame(X_pca,columns=['PCA1','PCA2']) #assigning own column names.
df_pca.head()


#Here 30 columns are made to 2 columns.
```

 Explained_variance_ratio

```python
explained_variance_ratio = pcaObject.explained_variance_ratio_ #It is Method

print(explained_variance_ratio *100)

#PCA1 column ---->44.2720

#PCA2 column ---->18.97
```

Missing Information

```python
100 - sum(explained_variance_ratio*100) #It is missing data.
```

```python
#Visualizing data for understanging breast _cancer 
```

```python
#Taking df['target'] and adding to the df_pca dataframe..

df_pca['target'] = df['target']
```

```python
df_pca.sample(6)
```

# maligant /Class-1 Data

```python
mask1 = df_pca['target'] == 1 #Using Boolean Mask we are filtering the data..
```

```python
malignant = df_pca[mask1]
```

```python
malignant.head(4) 
```

```python
mask2 = df_pca['target'] == 0 #filtering the begign data.
```

```python
beginin =df_pca[mask2]
beginin.head()
```

```python
#Visualizing the data..

plt.figure()
#-->Visualing the begin class 0 dataa.
plt.scatter(beginin['PCA1'],beginin['PCA2'],c='green',label='class 0')

#--->visualizing the maligant cancer calls 1 data.
plt.scatter(malignant['PCA1'],maligant['PCA2'],c='red',label='calss 1')

plt.xlabel("PCA1")

plt.ylabel("PCA2")

plt.title("PCA Component Analysis of Breast Cancer")

plt.legend()

plt.show()
```

# Task

Applying pca to the iris dataset..

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df= pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
df['target'] = iris['target'] 
```

```python
df.head(3)
```

```python
X = df.drop('target',axis=1)
```

```python
y = df['target']
```

```python
print(X.shape)

print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stand = StandardScaler()
```

```python
X_transform = stand.fit_transform(X)
```

```python
X.describe()
```

```python
pd.DataFrame(X_transform).describe()
```

```python
#Here it is changed the Values of the fEATURES.
```

```python
from sklearn.decomposition import PCA
```

```python
pca =PCA(n_components=2)
```

```python
X_pca=pca.fit_transform(X_transform)
```

```python
df1_pca = pd.DataFrame(X_pca,columns=['PCA1','PCA2'])
```

```python
df1_pca.head()
```

```python
df1_pca['target']=df['target']

df_pca.sample(3)
```

```python
#Visualizing data..
#To visualize the data we have to filter the data..
#Target ==0
#Target ==1
#Target ==2
```

```python
mask1 = df1_pca['target'] ==0

class_0 = df1_pca[mask1]

class_0['target'].unique()
```

```python
mask2 = df1_pca['target'] ==1

class_1 = df1_pca[mask2]

class_1['target'].unique()
```

```python
mask3 = df1_pca['target'] ==2

class_2 = df1_pca[mask3]

class_2['target'].unique()
```

```python
plt.figure()
plt.scatter(class_0['PCA1'],class_0['PCA2'],c="green",label='class 0')
plt.scatter(class_1['PCA1'],class_1['PCA2'],c="orange",label='class 1')
plt.scatter(class_2['PCA1'],class_2['PCA2'],c="blue",label='class 2')

plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title("PCA on Iris Dataset")

plt.legend()

plt.show()
```
